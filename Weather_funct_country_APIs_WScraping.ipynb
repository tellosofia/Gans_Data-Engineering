Now we'll write a function to get cities from defined countries and population. We need to write functions because that's what we'll use in AWS to connect automatically to the database, APIs, etc.


```python
# import the necessary libraries

import pandas as pd
import requests
import json
```

```python
# the other option with is to give directly the city, using city = instead of country
# Define the parameters to be passed into the url.
countries = ['ES', 'FR', 'PT', 'DE']                  # list of countries we want to get the cities from
min_population = "500000"                             # min population of the cities we want to get
limit = "5"                                           # number of cities we want to get per country
API_key = "you_API_key"  # your API key

cities = []                                           # empty list to store the cities

for i in range(len(countries)):                       # loop through the countries
  cities.append(requests.get(f"https://api.api-ninjas.com/v1/city?country={countries[i]}&min_population={min_population}&limit={limit}", headers={'X-Api-Key': API_key}).json())
```

```python
# to flatten the list
flattened_data = [item for sublist in cities for item in sublist]   # we use a for loop to flatten the list that we got from the API
cities_df = pd.DataFrame(flattened_data)                            # we convert the list into a dataframe
```

```python
def weather_per_city(countries, min_population, limit):
  """

  Args:
    countries: list of countries
    min_population: minimum population we'd like to search
    limit: max. number of cities we want per country

  Returns: dataframe with weather data

  """

  API_key = "your_API_key"    # insert your APIk

  cities = []                 # create an empty list to save values

  for i in range(len(countries)): # to loop through countries info to get cities with specified criteria
    cities.append(requests.get(f"https://api.api-ninjas.com/v1/city?country={countries[i]}&min_population={min_population}&limit={limit}", headers={'X-Api-Key': API_key}).json())

  flattened_data = [item for sublist in cities for item in sublist]
  cities_df = pd.DataFrame(flattened_data)

  latitude = list(cities_df['latitude'])
  longitude = list(cities_df['longitude'])

  API_key_b = 'your_API_key'

  weather = []

  for i in range(len(lat)): # loop through latitude and longitude to get weather
    weather.append(requests.get(f"http://api.openweathermap.org/data/2.5/forecast?lat={latitude[i]}&lon={longitude[i]}&appid={API_key_b}&units=metric").json())

  name = []
  date = []
  temp_min = []
  temp_max = []
  id = []
  weather_type = []
  weather_desc = []
  rain_P = []
  rain = []

  for i in range(len(weather)): # as we loop through the weather we get the values we need
    for j in range(len(weather[i]['list'])):
      name.append(weather[i]['city']['name'])
      date.append(weather[i]['list'][j]['dt_txt'])
      temp_min.append(weather[i]['list'][j]['main']['temp_min'])
      temp_max.append(weather[i]['list'][j]['main']['temp_max'])
      id.append(weather[i]['list'][j]['weather'][0]['id'])
      weather_type.append(weather[i]['list'][j]['weather'][0]['main'])
      weather_desc.append(weather[i]['list'][j]['weather'][0]['description'])
      rain_P.append(weather[i]['list'][j]['pop'])
      try:
        rain.append(weather[i]['list'][j]['rain']['3h'])
      except:
        rain.append(0)

  wt = list(zip(name, date, temp_min, temp_max, id, weather_type, weather_desc, rain_P, rain)) # and finally we create a dataframe
  weather_df = pd.DataFrame(wt, columns=['city', 'date', 'min_temp', 'max_temp', 'w_id', 'w_type', 'w_desc', 'rain_P', 'rain_mm'])

  return(weather_df) # the function returns the created dataframe
```
